```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel(r'C:\Users\ISUG\Documents\Qiyas\COVID-19-geographic-disbtribution-worldwide.xlsx')

# Extract African data only
africa_df = df[df['continentExp'] == 'Africa']

# Pick top 10 countries
top_countries = africa_df.groupby('countriesAndTerritories')['cases'].sum().nlargest(10).index
africa_top = africa_df[africa_df['countriesAndTerritories'].isin(top_countries)]

# Create pivot table - rows = countries, columns = months, values = total cases
pivot = africa_top.pivot_table(
    index='countriesAndTerritories',
    columns='month',
    values='cases',
    aggfunc='sum',
    fill_value=0
)

print(pivot)

# ================================
# HEATMAP
# ================================
plt.figure(figsize=(12, 6))

sns.heatmap(
    pivot,
    cmap='Reds',        # color scheme - white to dark red
    annot=True,         # show numbers inside each cell
    fmt='d',            # format numbers as integers
    linewidths=0.5,     # add grid lines between cells
    linecolor='white'   # grid line color
)

plt.title('COVID-19 Cases per African Country per Month')
plt.xlabel('Month')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel(r'C:\Users\ISUG\Documents\Qiyas\COVID-19-geographic-disbtribution-worldwide.xlsx')

# Extract African data only
africa_df = df[df['continentExp'] == 'Africa']

# Top 10 countries
top_countries = africa_df.groupby('countriesAndTerritories')['cases'].sum().nlargest(10).index
africa_top = africa_df[africa_df['countriesAndTerritories'].isin(top_countries)]

# Calculate death rate per country
country_stats = africa_top.groupby('countriesAndTerritories')[['cases', 'deaths']].sum()
country_stats['death_rate'] = (country_stats['deaths'] / country_stats['cases']) * 100

# Monthly totals
monthly = africa_top.groupby('month')[['cases', 'deaths']].sum()

# ================================
# DASHBOARD - 4 charts in one
# ================================
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('COVID-19 Africa Dashboard', fontsize=16, fontweight='bold')

# ================================
# Chart 1 - Bar chart: Total cases per country
# ================================
country_cases = country_stats['cases'].sort_values(ascending=False)
axes[0, 0].bar(country_cases.index, country_cases.values, color='tab:blue')
axes[0, 0].set_title('Total Cases per Country')
axes[0, 0].set_xlabel('Country')
axes[0, 0].set_ylabel('Total Cases')
axes[0, 0].tick_params(axis='x', rotation=90)

# ================================
# Chart 2 - Bar chart: Death rate per country
# ================================
death_rate = country_stats['death_rate'].sort_values(ascending=False)
axes[0, 1].bar(death_rate.index, death_rate.values, color='tab:red')
axes[0, 1].set_title('Death Rate per Country (%)')
axes[0, 1].set_xlabel('Country')
axes[0, 1].set_ylabel('Death Rate (%)')
axes[0, 1].tick_params(axis='x', rotation=90)

# ================================
# Chart 3 - Line chart: Monthly cases trend
# ================================
axes[1, 0].plot(monthly.index, monthly['cases'], marker='o', color='tab:green', label='Cases')
axes[1, 0].plot(monthly.index, monthly['deaths'], marker='o', color='tab:red', label='Deaths')
axes[1, 0].set_title('Monthly Cases and Deaths Trend')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Count')
axes[1, 0].legend()

# ================================
# Chart 4 - Heatmap: Cases per country per month
# ================================
pivot = africa_top.pivot_table(
    index='countriesAndTerritories',
    columns='month',
    values='cases',
    aggfunc='sum',
    fill_value=0
)
sns.heatmap(pivot, cmap='Reds', annot=True, fmt='d',
            linewidths=0.5, ax=axes[1, 1])
axes[1, 1].set_title('Cases per Country per Month')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel(r'C:\Users\ISUG\Documents\Qiyas\COVID-19-geographic-disbtribution-worldwide.xlsx')

# Extract African data only
africa_df = df[df['continentExp'] == 'Africa']

# Calculate total cases and deaths per country
country_stats = africa_df.groupby('countriesAndTerritories')[['cases', 'deaths']].sum()

# Calculate death rate
country_stats['death_rate'] = (country_stats['deaths'] / country_stats['cases']) * 100

# Sort by death rate highest to lowest
country_stats = country_stats.sort_values('death_rate', ascending=False)

print(country_stats, '\n')

# ================================
# BAR CHART - Death Rate per Country
# ================================
plt.figure(figsize=(14, 6))
plt.bar(country_stats.index, country_stats['death_rate'], color='tab:red')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Death Rate (%)')
plt.title('COVID-19 Death Rate per African Country')
plt.tight_layout()
plt.show()
```
