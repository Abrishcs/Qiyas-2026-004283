```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

headers = {"User-Agent": "Mozilla/5.0"}

# ================================
# ENGLISH CATEGORIES
# ================================
english_categories = {
    "Home"          : "https://borkena.com/",
    "Featured"      : "https://borkena.com/latest-ethiopian-news/",
    "Latest"        : "https://borkena.com/ethiopian-news-today/",
    "News"          : "https://borkena.com/ethiopia-news/",
    "Business"      : "https://borkena.com/ethiopia-business-news/",
    "Opinion"       : "https://borkena.com/ethiopian-news-and-opinion/",
    "Sport"         : "https://borkena.com/sport/",
    "Entertainment" : "https://borkena.com/entertainment-ethiopian-music-drama-show/",
    "Restaurant"    : "https://borkena.com/ethiopian-restaurant/",
    "Travel"        : "https://borkena.com/travel/",
    "Video"         : "https://borkena.com/ethiopian-videos/",
}

# ================================
# AMHARIC CATEGORIES
# ================================
amharic_categories = {
    "Amharic_News" : "https://amharic.borkena.com/",
}

# ================================
# CREATE FOLDERS
# ================================
os.makedirs("borkena_data/English", exist_ok=True)
os.makedirs("borkena_data/Amharic", exist_ok=True)

# ================================
# FUNCTION TO SCRAPE ONE PAGE
# ================================
def scrape_page(url, category, language):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    articles = soup.find_all("h3")

    data = []
    for article in articles:
        link_tag = article.find("a")
        if link_tag:
            title = link_tag.text.strip()
            link  = link_tag["href"]
            if title != "":
                data.append({
                    "Title"    : title,
                    "Link"     : link,
                    "Category" : category,
                    "Language" : language
                })
    return data

# ================================
# SCRAPE ENGLISH PAGES
# ================================
print("=" * 40)
print("SCRAPING ENGLISH PAGES...")
print("=" * 40)

all_english = []

for category, url in english_categories.items():
    data = scrape_page(url, category, "English")
    df = pd.DataFrame(data)
    filename = f"borkena_data/English/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"✅ {category} → {len(data)} articles")
    all_english.extend(data)

# Save all English combined
english_df = pd.DataFrame(all_english).reset_index(drop=True)
english_df.to_excel("borkena_data/English/ALL_English.xlsx", index=False)
print(f"\nTotal English articles: {len(all_english)}")

# ================================
# SCRAPE AMHARIC PAGES
# ================================
print("\n" + "=" * 40)
print("SCRAPING AMHARIC PAGES...")
print("=" * 40)

all_amharic = []

for category, url in amharic_categories.items():
    data = scrape_page(url, category, "Amharic")
    df = pd.DataFrame(data)
    filename = f"borkena_data/Amharic/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"✅ {category} → {len(data)} articles")
    all_amharic.extend(data)

# Save all Amharic combined
amharic_df = pd.DataFrame(all_amharic).reset_index(drop=True)
amharic_df.to_excel("borkena_data/Amharic/ALL_Amharic.xlsx", index=False)
print(f"\nTotal Amharic articles: {len(all_amharic)}")

# ================================
# COMBINE EVERYTHING
# ================================
print("\n" + "=" * 40)
print("COMBINING EVERYTHING...")
print("=" * 40)

all_df = pd.concat([english_df, amharic_df]).reset_index(drop=True)
all_df.to_excel("borkena_data/ALL_combined.xlsx", index=False)

print(f"\nGrand Total articles: {len(all_df)}")
print("\nArticles per category:")
print(all_df["Category"].value_counts())
print("\nArticles per language:")
print(all_df["Language"].value_counts())
print("\n✅ ALL DONE! Check borkena_data folder!")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

headers = {"User-Agent": "Mozilla/5.0"}

# ================================
# ENGLISH CATEGORIES
# ================================
english_categories = {
    "Home"          : "https://borkena.com/",
    "Featured"      : "https://borkena.com/latest-ethiopian-news/",
    "Latest"        : "https://borkena.com/ethiopian-news-today/",
    "News"          : "https://borkena.com/ethiopia-news/",
    "Business"      : "https://borkena.com/ethiopia-business-news/",
    "Opinion"       : "https://borkena.com/ethiopian-news-and-opinion/",
    "Sport"         : "https://borkena.com/sport/",
    "Entertainment" : "https://borkena.com/entertainment-ethiopian-music-drama-show/",
    "Restaurant"    : "https://borkena.com/ethiopian-restaurant/",
    "Travel"        : "https://borkena.com/travel/",
    "Video"         : "https://borkena.com/ethiopian-videos/",
}

# ================================
# AMHARIC CATEGORIES
# ================================
amharic_categories = {
    "Amharic_News" : "https://amharic.borkena.com/",
}

# ================================
# CREATE FOLDERS
# ================================
os.makedirs("borkena_data/English", exist_ok=True)
os.makedirs("borkena_data/Amharic", exist_ok=True)

# ================================
# UPDATED SCRAPE FUNCTION
# ================================
def scrape_page(url, category, language):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all article blocks
    articles = soup.find_all("h3")

    data = []
    for article in articles:
        link_tag = article.find("a")
        if link_tag:
            title = link_tag.text.strip()
            link  = link_tag["href"]

            if title == "":
                continue

            # ── Find the parent container of article ──
            parent = article.find_parent()

            # ── AUTHOR ──
            try:
                author = parent.find("a", href=lambda h: h and "/author/" in h)
                author = author.text.strip() if author else "Unknown"
            except:
                author = "Unknown"

            # ── DATE ──
            try:
                date = parent.find("span", class_=lambda c: c and "date" in c.lower())
                if not date:
                    # Try finding any text that looks like a date
                    date = parent.find(
                        lambda tag: tag.name in ["span", "p", "div"] 
                        and any(month in tag.text for month in 
                               ["January","February","March","April",
                                "May","June","July","August",
                                "September","October","November","December"])
                    )
                date = date.text.strip() if date else "Unknown"
            except:
                date = "Unknown"

            # ── IMAGE ──
            try:
                img = parent.find("img")
                image_url = img["src"] if img and "src" in img.attrs else "No Image"
            except:
                image_url = "No Image"

            data.append({
                "Title"     : title,
                "Author"    : author,
                "Date"      : date,
                "Image URL" : image_url,
                "Link"      : link,
                "Category"  : category,
                "Language"  : language
            })

    return data

# ================================
# SCRAPE ENGLISH PAGES
# ================================
print("=" * 40)
print("SCRAPING ENGLISH PAGES...")
print("=" * 40)

all_english = []

for category, url in english_categories.items():
    data = scrape_page(url, category, "English")
    df = pd.DataFrame(data)
    filename = f"borkena_data/English/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"✅ {category} → {len(data)} articles")
    all_english.extend(data)

english_df = pd.DataFrame(all_english).reset_index(drop=True)
english_df.to_excel("borkena_data/English/ALL_English.xlsx", index=False)
print(f"\nTotal English articles: {len(all_english)}")

# ================================
# SCRAPE AMHARIC PAGES
# ================================
print("\n" + "=" * 40)
print("SCRAPING AMHARIC PAGES...")
print("=" * 40)

all_amharic = []

for category, url in amharic_categories.items():
    data = scrape_page(url, category, "Amharic")
    df = pd.DataFrame(data)
    filename = f"borkena_data/Amharic/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"✅ {category} → {len(data)} articles")
    all_amharic.extend(data)

amharic_df = pd.DataFrame(all_amharic).reset_index(drop=True)
amharic_df.to_excel("borkena_data/Amharic/ALL_Amharic.xlsx", index=False)
print(f"\nTotal Amharic articles: {len(all_amharic)}")

# ================================
# COMBINE EVERYTHING
# ================================
print("\n" + "=" * 40)
print("COMBINING EVERYTHING...")
print("=" * 40)

all_df = pd.concat([english_df, amharic_df]).reset_index(drop=True)
all_df.to_excel("borkena_data/ALL_combined.xlsx", index=False)

print(f"\nGrand Total: {len(all_df)} articles")
print("\nPer category:")
print(all_df["Category"].value_counts())
print("\nPer language:")
print(all_df["Language"].value_counts())
print("\n✅ ALL DONE!")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import time

headers = {"User-Agent": "Mozilla/5.0"}

# ================================
# SET MAX PAGES PER CATEGORY
# ================================
# Change these numbers as you want!
# Use small numbers for testing
# Use big numbers for full scraping

category_pages = {
    "Home"          : 3,
    "Featured"      : 5,
    "Latest"        : 10,   # has 276 pages! use 10 for demo
    "News"          : 10,
    "Business"      : 5,
    "Opinion"       : 5,
    "Sport"         : 5,
    "Entertainment" : 5,
    "Restaurant"    : 3,
    "Travel"        : 3,
    "Video"         : 3,
    "Amharic_News"  : 5,
}

# ================================
# CATEGORY URLS
# ================================
english_categories = {
    "Home"          : "https://borkena.com/",
    "Featured"      : "https://borkena.com/latest-ethiopian-news/",
    "Latest"        : "https://borkena.com/ethiopian-news-today/",
    "News"          : "https://borkena.com/ethiopia-news/",
    "Business"      : "https://borkena.com/ethiopia-business-news/",
    "Opinion"       : "https://borkena.com/ethiopian-news-and-opinion/",
    "Sport"         : "https://borkena.com/sport/",
    "Entertainment" : "https://borkena.com/entertainment-ethiopian-music-drama-show/",
    "Restaurant"    : "https://borkena.com/ethiopian-restaurant/",
    "Travel"        : "https://borkena.com/travel/",
    "Video"         : "https://borkena.com/ethiopian-videos/",
}

amharic_categories = {
    "Amharic_News" : "https://amharic.borkena.com/",
}

# ================================
# CREATE FOLDERS
# ================================
os.makedirs("borkena_data/English", exist_ok=True)
os.makedirs("borkena_data/Amharic", exist_ok=True)

# ================================
# SCRAPE ONE PAGE FUNCTION
# ================================
def scrape_one_page(url, category, language):
    try:
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.content, "html.parser")
        articles = soup.find_all("h3")

        data = []
        for article in articles:
            link_tag = article.find("a")
            if link_tag:
                title = link_tag.text.strip()
                link  = link_tag["href"]

                if title == "":
                    continue

                parent = article.find_parent()

                # ── AUTHOR ──
                try:
                    author = parent.find(
                        "a", href=lambda h: h and "/author/" in h)
                    author = author.text.strip() if author else "Unknown"
                except:
                    author = "Unknown"

                # ── DATE ──
                try:
                    date = parent.find(
                        lambda tag: tag.name in ["span","p","div","time"]
                        and any(m in tag.text for m in
                               ["January","February","March","April",
                                "May","June","July","August",
                                "September","October","November","December"])
                    )
                    date = date.text.strip() if date else "Unknown"
                except:
                    date = "Unknown"

                # ── IMAGE ──
                try:
                    img = parent.find("img")
                    if img:
                        image_url = (img.get("data-src") or
                                    img.get("data-lazy-src") or
                                    img.get("src") or
                                    "No Image")
                    else:
                        image_url = "No Image"
                except:
                    image_url = "No Image"

                data.append({
                    "Title"     : title,
                    "Author"    : author,
                    "Date"      : date,
                    "Image URL" : image_url,
                    "Link"      : link,
                    "Category"  : category,
                    "Language"  : language
                })

        return data, soup

    except Exception as e:
        print(f"   ❌ Error: {e}")
        return [], None

# ================================
# SCRAPE ALL PAGES FUNCTION
# ================================
# ================================
# FIXED SCRAPE ALL PAGES FUNCTION
# ================================
def scrape_all_pages(base_url, category, language):
    max_pages = category_pages.get(category, 5)
    all_data  = []

    for page in range(1, max_pages + 1):

        # Build URL for each page
        if page == 1:
            url = base_url
        else:
            url = base_url.rstrip("/") + f"/page/{page}/"

        print(f"   📄 Page {page}/{max_pages} → scraping...")

        # Scrape the page
        data, soup = scrape_one_page(url, category, language)

        # If no articles found → stop!
        if len(data) == 0:
            print(f"   ⚠️  No more articles → stopping at page {page}")
            break

        all_data.extend(data)
        print(f"   ✅ Page {page} → {len(data)} articles found")

        # Wait between pages → be polite!
        time.sleep(1)

    return all_data

# ================================
# SCRAPE ENGLISH PAGES
# ================================
print("=" * 45)
print("  SCRAPING ENGLISH PAGES...")
print("=" * 45)

all_english = []

for category, url in english_categories.items():
    print(f"\n🔍 Category: {category}")
    print(f"   Max pages: {category_pages[category]}")

    data = scrape_all_pages(url, category, "English")

    df = pd.DataFrame(data)
    filename = f"borkena_data/English/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"💾 Saved → {len(data)} articles total")
    all_english.extend(data)

english_df = pd.DataFrame(all_english).reset_index(drop=True)
english_df.to_excel("borkena_data/English/ALL_English.xlsx", index=False)
print(f"\n✅ Total English: {len(all_english)} articles")

# ================================
# SCRAPE AMHARIC PAGES
# ================================
print("\n" + "=" * 45)
print("  SCRAPING AMHARIC PAGES...")
print("=" * 45)

all_amharic = []

for category, url in amharic_categories.items():
    print(f"\n🔍 Category: {category}")
    print(f"   Max pages: {category_pages[category]}")

    data = scrape_all_pages(url, category, "Amharic")

    df = pd.DataFrame(data)
    filename = f"borkena_data/Amharic/{category}.xlsx"
    df.to_excel(filename, index=False)
    print(f"💾 Saved → {len(data)} articles total")
    all_amharic.extend(data)

amharic_df = pd.DataFrame(all_amharic).reset_index(drop=True)
amharic_df.to_excel("borkena_data/Amharic/ALL_Amharic.xlsx", index=False)
print(f"\n✅ Total Amharic: {len(all_amharic)} articles")

# ================================
# COMBINE EVERYTHING
# ================================
print("\n" + "=" * 45)
print("  COMBINING EVERYTHING...")
print("=" * 45)

all_df = pd.concat([english_df, amharic_df]).reset_index(drop=True)
all_df.to_excel("borkena_data/ALL_combined.xlsx", index=False)

print(f"\n📊 FINAL RESULTS:")
print(f"   English  : {len(all_english)} articles")
print(f"   Amharic  : {len(all_amharic)} articles")
print(f"   TOTAL    : {len(all_df)} articles")
print(f"\nPer category:")
print(all_df["Category"].value_counts().to_string())
print("\n✅ ALL DONE! Check borkena_data folder!")
```
