```python
import requests
from bs4 import BeautifulSoup

html_doc = requests.get("https://www.worldometers.info/coronavirus/")
soup = BeautifulSoup(html_doc.content, "html.parser")

table = soup.find("table")
tbody = table.find("tbody")
rows = tbody.find_all("tr")

f = open("countries.txt", "w", encoding="utf-8")

for row in rows:
    cols = row.find_all("td")

    if len(cols) > 1:
        country = cols[1].text.strip()
        f.write(country + "\n")

f.close()

print("Data saved successfully")
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

html_doc = requests.get("https://www.worldometers.info/coronavirus/", headers={"User-Agent": "Mozilla/5.0"})
soup = BeautifulSoup(html_doc.content, "html.parser")
table = soup.find("table")
tbody = table.find("tbody")
rows = tbody.find_all("tr")

# Create Excel file
file = openpyxl.Workbook()
sheet = file.active

# Write header row
sheet.append(["Country", "Total Cases", "Total Deaths", "Total Recovered", "Active Cases",
              "Serious/Critical", "Cases/1M Pop", "Deaths/1M Pop", "Total Tests",
              "Tests/1M Pop", "Population", "Continent"])

# Write data rows
for row in rows[8:]:
    cols = row.find_all("td")

    if len(cols) > 5:
        country          = cols[1].text.strip()
        total_cases      = cols[2].text.strip().replace(",", "")
        total_deaths     = cols[4].text.strip().replace(",", "")
        total_recovered  = cols[6].text.strip().replace(",", "").replace("+", "")
        active_cases     = cols[8].text.strip().replace(",", "")
        serious_critical = cols[9].text.strip().replace(",", "")
        cases_1m         = cols[10].text.strip().replace(",", "")
        deaths_1m        = cols[11].text.strip().replace(",", "")
        total_tests      = cols[12].text.strip().replace(",", "")
        tests_1m         = cols[13].text.strip().replace(",", "")
        population       = cols[14].text.strip().replace(",", "")
        continent        = cols[15].text.strip()

        if country == "":
            continue

        sheet.append([country, total_cases, total_deaths, total_recovered, active_cases,
                      serious_critical, cases_1m, deaths_1m, total_tests,
                      tests_1m, population, continent])

# Fix column widths
sheet.column_dimensions["A"].width = 20  # Country
sheet.column_dimensions["B"].width = 15  # Total Cases
sheet.column_dimensions["C"].width = 15  # Total Deaths
sheet.column_dimensions["D"].width = 18  # Total Recovered
sheet.column_dimensions["E"].width = 15  # Active Cases
sheet.column_dimensions["F"].width = 18  # Serious/Critical
sheet.column_dimensions["G"].width = 15  # Cases/1M Pop
sheet.column_dimensions["H"].width = 15  # Deaths/1M Pop
sheet.column_dimensions["I"].width = 15  # Total Tests
sheet.column_dimensions["J"].width = 15  # Tests/1M Pop
sheet.column_dimensions["K"].width = 15  # Population
sheet.column_dimensions["L"].width = 15  # Continent

file.save("countries6.xlsx")
print("Data saved successfully!")
```
