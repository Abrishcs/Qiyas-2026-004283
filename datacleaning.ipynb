```python
import pandas as pd

# Create the dataset
data = {
    'ID'    : ['Eth-101', 'Eth-102', 'Eth-103', 'Eth-104', 'Eth-105'],
    'Name'  : ['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse'],
    'Age'   : [25, None, 29, 33, None],
    'Salary': [55000, 48000, None, 62000, None]
}

df = pd.DataFrame(data)

# Create a column-based flag
for column in df.columns:
    if df[column].isnull().any():
        df[f'{column}_Flag'] = df[column].isna().apply(
            lambda x: 'Missing' if x else 'Not Missing'
        )

# Create a row-based flag
df['Missing_Flag'] = df.isnull().any(axis=1).apply(
    lambda x: 'Missing' if x else 'Not Missing'
)

print(df)
```
