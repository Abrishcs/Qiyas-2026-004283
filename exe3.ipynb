```python
from dash import Dash, dcc, html, Input, Output
import pandas as pd

# ================================
# LOAD DATA
# ================================
df = pd.read_excel("borkena_data/ALL_combined.xlsx")

# Clean date column
df['Date'] = pd.to_datetime(df['Date'], format='mixed', errors='coerce')
df = df.dropna(subset=['Date'])
df['Date_str'] = df['Date'].dt.strftime('%Y-%m-%d')

# Get unique values for dropdowns
languages  = sorted(df['Language'].unique().tolist())
dates      = sorted(df['Date_str'].unique().tolist(), reverse=True)

# ================================
# CREATE APP
# ================================
app = Dash(__name__)

app.layout = html.Div([

    # ── HEADER ──
    html.H1('📰 Borkena News Dashboard',
            style={'textAlign': 'center',
                   'color'    : 'darkblue',
                   'padding'  : '20px'}),

    # ── DROPDOWNS ROW ──
    html.Div([

        # Language Dropdown
        html.Div([
            html.Label('🌐 Select Language:',
                       style={'fontWeight': 'bold'}),
            dcc.Dropdown(
                id='language-dropdown',
                options=[{'label': l, 'value': l} for l in languages],
                value='English',
                clearable=False,
            ),
        ], style={'width': '30%', 'display': 'inline-block',
                  'marginRight': '3%'}),

        # Category Dropdown
        html.Div([
            html.Label('📂 Select Category:',
                       style={'fontWeight': 'bold'}),
            dcc.Dropdown(
                id='category-dropdown',
                options=[],
                value=None,
                clearable=False,
            ),
        ], style={'width': '30%', 'display': 'inline-block',
                  'marginRight': '3%'}),

        # Date Dropdown
        html.Div([
            html.Label('📅 Select Date:',
                       style={'fontWeight': 'bold'}),
            dcc.Dropdown(
                id='date-dropdown',
                options=[{'label': d, 'value': d} for d in dates],
                value=dates[0],
                clearable=False,
            ),
        ], style={'width': '30%', 'display': 'inline-block'}),

    ], style={'padding': '20px', 'backgroundColor': '#f8f9fa',
              'borderRadius': '10px', 'margin': '20px'}),

    # ── RESULTS COUNT ──
    html.Div(id='results-count',
             style={'padding'   : '10px 20px',
                    'fontSize'  : '16px',
                    'color'     : 'gray'}),

    html.Hr(),

    # ── ARTICLES TABLE ──
    html.Div(id='articles-table',
             style={'padding': '20px'}),

])

# ================================
# CALLBACK 1 - Update category dropdown
# when language changes
# ================================
@app.callback(
    Output('category-dropdown', 'options'),
    Output('category-dropdown', 'value'),
    Input('language-dropdown',  'value')
)
def update_categories(selected_language):
    filtered   = df[df['Language'] == selected_language]
    categories = sorted(filtered['Category'].unique().tolist())
    options    = [{'label': c, 'value': c} for c in categories]
    return options, categories[0]

# ================================
# CALLBACK 2 - Update articles table
# ================================
@app.callback(
    Output('articles-table', 'children'),
    Output('results-count',  'children'),
    Input('language-dropdown',  'value'),
    Input('category-dropdown',  'value'),
    Input('date-dropdown',      'value')
)
def update_table(selected_language, selected_category, selected_date):

    if selected_category is None:
        return [], "No results found"

    # Filter data
    filtered = df[
        (df['Language']  == selected_language) &
        (df['Category']  == selected_category) &
        (df['Date_str']  == selected_date)
    ]

    # Results count
    count = f"📊 Found {len(filtered)} articles"

    if len(filtered) == 0:
        return html.P("No articles found for this selection.",
                      style={'color': 'red', 'fontSize': '16px'}), count

    # Build table rows
    rows = []

    # Table header
    rows.append(
        html.Tr([
            html.Th("Title",    style={'width': '50%', 'padding': '10px',
                                       'backgroundColor': 'darkblue',
                                       'color': 'white'}),
            html.Th("Author",   style={'width': '15%', 'padding': '10px',
                                       'backgroundColor': 'darkblue',
                                       'color': 'white'}),
            html.Th("Date",     style={'width': '15%', 'padding': '10px',
                                       'backgroundColor': 'darkblue',
                                       'color': 'white'}),
            html.Th("Category", style={'width': '10%', 'padding': '10px',
                                       'backgroundColor': 'darkblue',
                                       'color': 'white'}),
            html.Th("Link",     style={'width': '10%', 'padding': '10px',
                                       'backgroundColor': 'darkblue',
                                       'color': 'white'}),
        ])
    )

    # Table rows
    for i, row in filtered.iterrows():
        rows.append(
            html.Tr([
                html.Td(row['Title'],    style={'padding': '10px'}),
                html.Td(row['Author'],   style={'padding': '10px'}),
                html.Td(row['Date_str'], style={'padding': '10px'}),
                html.Td(row['Category'], style={'padding': '10px'}),
                html.Td(
                    html.A("Read →", href=row['Link'], target="_blank",
                           style={'color': 'darkblue'}),
                    style={'padding': '10px'}
                ),
            ], style={'backgroundColor': '#f2f2f2' if i % 2 == 0 else 'white'})
        )

    table = html.Table(rows, style={
        'width'          : '100%',
        'borderCollapse' : 'collapse',
        'border'         : '1px solid #ddd',
        'fontSize'       : '14px'
    })

    return table, count

if __name__ == '__main__':
    app.run(debug=True)
```
